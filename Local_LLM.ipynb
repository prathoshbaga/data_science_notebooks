```python
import os
from langchain_community.llms import Ollama
from crewai import Agent, Task, Crew, Process

model = Ollama(model="llama3")

recommender = Agent(
    role = "AI/ML code generator and model imporvement suggestion for pytorch",
    goal = "Based on the question which the user asks, generate relevant code if asked for code generation or provide tips to improve the machine learing/Artificial Intelligence model for pytorch. no matter what, be very concise",
    backstory = "You are an AI assitant whose only job is to generate AI/ML code or suggest model imporvementmfor pytorch based on user query.",
    verbose = True,
    allow_delegation = False,
    llm = model
)

recommendation = Task(
    description = "generate AI/ML code or suggest model imporvements for pytorch based on user query {query}",
    agent = recommender,
    expected_output = "a very concise response to generation AI/ML code or  model imporvements suggestion for pytorch based on user query.",
    input_variables = ["query"]
)

crew = Crew(
    agents = [recommender],
    tasks = [recommendation],
    verbose=2,
    process=Process.sequential,
    memory=True,
    cache=True,
)

while True:
    user_input = input("Ask me anything : ")
    if user_input.lower() == "quit":
      print("!!!! Exiting Recommender !!!!")
      break
    else:
      result = crew.kickoff(inputs={'query': user_input})
      print(result)
```
