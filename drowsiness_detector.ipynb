```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import torch
%matplotlib inline
```

```python
model = torch.hub.load("ultralytics/yolov5","yolov5s", pretrained=True)
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
  ret,frame = cap.read()

  # make detection
  result = model(frame)
  cv2.imshow("YOLO",np.squeeze(result.render()))

  if cv2.waitKey(10) & 0xFF ==ord("q"):
    break
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join(r"C:\Users\PBaga\Desktop\workspace\data-science\data","images")
labels = ["Awake", "Drowsy"]
number_of_imgs = 20
print(IMAGES_PATH)
```

```python
cap = cv2.VideoCapture(0)
#loop throuch labels
for label in labels:
    print(f"Capturing images for {label}")
    time.sleep(5)

    #loop through images
    for img in range(number_of_imgs):
        print(f"Capturing images for {label}, image number {img}"+".jpg")

        ret, frame = cap.read()

        imgname = os.path.join(IMAGES_PATH,label+"."+str(uuid.uuid1()))
        print(f"Capturing {imgname}")

        status = cv2.imwrite(imgname,frame)
        if os.access(imgname, os.W_OK):
            if status:
                print("Image saved successfully")
            else:
                print("Image not saved")
        else:
            print("No permission")

        cv2.imshow("Image Collection", frame)
        time.sleep(2)

        if cv2.waitKey(10) & 0xFF ==ord("q"):
            break
cap.release()
cv2.destroyAllWindows()

```

```python
import os
import shutil

# Set the paths for the folders
txt_folder = r'C:\Users\PBaga\Desktop\workspace\data-science\data\OpenLabeling\main\output\YOLO_darknet'
image_folder = r'C:\Users\PBaga\Desktop\workspace\data-science\data\OpenLabeling\main\input\drowsy_mp4'
#C:\Users\PBaga\Desktop\workspace\data-science\data\OpenLabeling\main\input\drowsy_mp4
output_folder = r'C:\Users\PBaga\Desktop\workspace\data-science\data\OpenLabeling\main\output\YOLO_darknet\drowsy_mp4'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through the files in the txt_folder
for filename in os.listdir(txt_folder):
    if filename.endswith('.txt'):
        # Get the base name of the file (without the .txt extension)
        base_name = os.path.splitext(filename)[0]

        # Check if there's a corresponding image file in the image_folder
        image_path = os.path.join(image_folder, base_name + '.jpg')
        if os.path.exists(image_path):
            # Move the image file to the output_folder
            output_image_path = os.path.join(output_folder, filename.replace('.txt', '.jpg'))
            shutil.move(image_path, output_image_path)
            print(f'Moved {base_name}.jpg to {output_folder}')
        else:
            print(f'No image found for {base_name}.txt')
```
